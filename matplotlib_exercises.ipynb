```python

import matplotlib.pyplot as plt  


import numpy as np
import math
from random import randint
import random
import pandas as pd
```

1. Plot y = x ^ 2 -x + 2 and add 0,0 origin text

```python
x = range(-70,70)
y = [x**2 - x + 2 for x in x]

plt.scatter(x,y)
plt.text(0,0, 'Origin')
plt.show()
```
